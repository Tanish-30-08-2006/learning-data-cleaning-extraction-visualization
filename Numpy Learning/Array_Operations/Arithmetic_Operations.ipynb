```python
import numpy as np
```

```python
#-------------ARITHMETIC OPEARTIONS-----------#

a1 = np.arange(1,6)
a2 = np.arange(6,11)
print( a1 , a2 )
```

```python
# while adding,subtracting size must be same , common usecase in pandas in any dataframe null values must be filled by something 
# so opearations can be performed

print(a1+a2)
print(a1-a2)
print(a1*a2)
print(a1/a2)
print(a1//a2) # a1//a2 is floor divison , integer part of output is considered
print(a1**a2) # a1**a2 means power function (element of a1) to the power (element of a2)
```

```python
#-------------ON 2D MATRICES-------------#
arr1 = np.arange(1,26).reshape(5,5)
print(arr1)

print(arr1+10)
print(arr1*2)
```
