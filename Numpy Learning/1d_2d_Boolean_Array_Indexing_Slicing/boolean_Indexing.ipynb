```python
import numpy as np
```

```python
#---------------BOOLEAN INDEXING-------------3

#  TO GET EVEN NUMBERS FROM ARRAY WE CANT SLICE IT 

arr = np.arange(1,11)
print(arr)

bool_index = arr%2==0
print(bool_index)
print(arr[bool_index])
```

```python
# TO GET ODD NUMBERS

print(arr)
bool_index = arr%2 !=0
print(bool_index)
print(arr[bool_index])
```
