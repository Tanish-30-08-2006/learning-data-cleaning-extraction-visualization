```python
import numpy as np
```

```python
#-----------------------GENERATE 1D ARRAY--------------#

arr = np.arange(1,11) # to create array of 1 to N-1 just like for loops 
print(arr)
arr1 = np.zeros(6) 
print(arr1)
```

```python
#--------------------GENERATE 2D ARRAY OR MATRICES----------------------------#
arr = np.zeros((3,3))
print(arr)

arr1 = np.ones((2,3))
print(arr1)

arr2 = np.zeros((6,10))
print(arr2)
```

```python
#-----------------------LINSPACE FUNCTION---------------------#

# press shift tab to see what any function does..
# np.linspace(start, end , no. of evenly spaced numbers we want)  ---> returns evenly spaced numbers over a specified interval

arr = np.linspace(1,5,2)
print(arr)

arr = np.linspace(1,5,3)
print(arr)

arr = np.linspace(1,5,5)
print(arr)

arr = np.linspace(1,5,10)
print(arr)

```
