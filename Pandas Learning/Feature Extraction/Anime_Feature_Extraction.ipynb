```python
#----------------FUNCTION DEBUGGING TO CALCULATE TOTAL PERIOD---------------#

def get_total_period(row):

   
    # month difference : 
    start_month_name = str(row['Start_Month']).lower()
    end_month_name   = str(row['End_Month']).lower()
    print(f"Start Month Name  : {start_month_name} || End Month Name  : {end_month_name} ")
    
    # mapping to integers to calculate difference : 
    
    if start_month_name in month_map : 
        start_month_integer = month_map[start_month_name]
    else : 
        start_month_integer = 1

    if end_month_name in month_map : 
        end_month_integer = month_map[end_month_name]
    else :
        end_month_integer = 1

    month_difference = end_month_integer - start_month_integer 

    # Year difference : 
    start_year = int(row['Start_Year'])
    end_year   = int(row['End_Year'])
    print(f"Start Year : {start_year} || End Year : {end_year} ")
    

    # calculate period :
    month_diff = end_month_integer - start_month_integer
    year_diff  = end_year - start_year
    print(f"Month Diff : {month_diff} || Year Diff :  {year_diff} ")
    

    period = int( (12*year_diff) +(month_diff) )
    print("Period : " , period)
    print("\n")

    if (period)<=0 :
        return 1
    
    return period

# ERROR IN THIS : display(df.loc[[1,2,3]].apply(get_total_period))

# HERE , previously we did df.loc[[1,2,3]].apply('Title') which is by default 
# applied on columns ( axis = 0 ) and here we want to calculate period by using
# attributes of every row(start year end year etc) so we have to specially write axis=1 for rows..

display(df.head(10).apply(get_total_period,axis=1))
    
```

```python
#----------------APPLYING FUNCTION-----------------#

# writing new funciton without debugging prints

def get_total_period(row):

   
    # month difference : 
    start_month_name = str(row['Start_Month']).lower()
    end_month_name   = str(row['End_Month']).lower()
    
    # mapping to integers to calculate difference : 
    
    if start_month_name in month_map : 
        start_month_integer = month_map[start_month_name]
    else : 
        start_month_integer = 1

    if end_month_name in month_map : 
        end_month_integer = month_map[end_month_name]
    else :
        end_month_integer = 1

    month_difference = end_month_integer - start_month_integer 

    # Year difference : 
    start_year = int(row['Start_Year'])
    end_year   = int(row['End_Year'])
    
    

    # calculate period :
    month_diff = end_month_integer - start_month_integer
    year_diff  = end_year - start_year
    

    period = int( (12*year_diff) +(month_diff) )
    

    if (period)<=0 :
        return 1
    
    return period

df['Period_Months'] = df.apply(get_total_period,axis=1)
```

```python
#-------------------------------Q:8------------------------------------------#

#--------FUNCTION DEBUGGING------------#

# To calclate pace of anime release ( we do (episode count) / period_months) that gives us which anime was most actively releasing
# episodes as pace gives episode released per month...

def calculate_pace(row):
    
    episode_count  = row['Episode_Count']
    period_months  = row['Period_Months']
    print(f"Episode count : {episode_count} || Period in Months { period_months} ")

    pace = float(episode_count)/(period_months)
    print("Pace : ",pace)
    print("\n")

    return pace

df.loc[[1,2,3]].apply(calculate_pace,axis=1) 
# by default axis=0 that is column gets passed here we have to pass the row to access 
# row attributes which ae episode_count and period_months ||  (axis=1) for rows
```

```python
# -------------------APPLTYING FUNCTION------------------#

def calculate_pace(row):
    
    episode_count  = row['Episode_Count']
    period_months  = row['Period_Months']
    
    pace = float((episode_count)/(period_months))

    return round(pace,2) # round(pace,2) limits the pace to 2 decimal digits ...

df['Pace'] = df.apply(calculate_pace,axis=1) 
```

```python
def extract_type(title):
    # Convert to string and lowercase
    t = str(title).lower()
    
    # THE FIX: Replace ALL weird whitespace (\n, \t, \xa0) with a standard space
    # This turns "O\nVA" or "O\tVA" into "o va"
   
    t = " ".join(t.split())

    # 3. Priority Check: Check for OVA variations FIRST
    if 'ova' in t or 'o va' in t or 'o.v.a' in t:
        return 'OVA'
    elif 'tv' in t: 
        return 'TV'
    elif 'movie' in t:
        return 'MOVIE'
    elif 'special' in t:
        return 'SPECIAL'
    else:
        return 'OTHER'

# Apply to the whole DataFrame
df['Type'] = df['Title'].apply(extract_type)

# Let's verify Row 50 (Index 49)
display(df.loc[[49]][['Title' ,  'Type']])

display(df)
```

```python
#---------------------------------SCORE HISTOGRAM-------------------------------#

# WHAT IS HISTOGRAM...
# Histogram looks like a bar chart but its different from that ...
# It takes a continous range of numbers ( score from 0-10) and chops them in bins( 8-8.2 , 8.2-8.4, ...) and calculates how many animes lie in that bin ( i.e score)

# WHEN DO WE USE HISTOGRAM...
# We use histogram when we want to find the shape of the data we have 
# is it a NORMAL DISTRIBUTION ( where most of our data is in the middle and going away from middle the number gets lower which is called BELL CURVE)
# is it a SKEWED DISTRIBUTION ( where most of our data lies on one side for eg : are there way more high scoring animes than low scoring ones in top 50 list)
# are there any outliers ( animes which are far away from the rest in the group)

# THE INSIGHT WE GET..
# Instead of just knowing the "Average Score" is 8.5, a Histogram tells you if most animes are exactly 8.5, or if they are spread out wildly between 7.0 and 9.5.
```

```python

plt.figure(figsize=(10,7))

#  'bins' defines how many bars we want
counts, bin_edges, patches = plt.hist(df['Score'], bins=20 , color='skyblue', edgecolor='black',alpha=0.7)

# adding the grid 
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.grid(axis='x', linestyle='--', alpha=0.4)

# rotation so 8.71 of firts bin doesnt overlaps with 8.72 of starting of second bin 
# adding bin edges so we know each bar starts and ends at what value of score so we can get better insights 
# changed bins from 4 decimals to 2 decimals using formatted labels function , bins_edges give us numeric value and formatted_labels convert it into 2 decimal string..
formatted_labels = [f"{x:.2f}" for x in bin_edges]
plt.xticks(bin_edges,formatted_labels, rotation=45, fontsize=10)

# padding means adding space between title and graph box 
plt.title('Surgical Score Distribution: Bin-Aligned Analysis', fontsize=16, pad=15)
plt.xlabel('Exact Score Boundaries (Bin Edges)', fontsize=12)
plt.ylabel('Frequency (Count of Animes)', fontsize=12)



# addimg meam score and median score lines parallel to y axis i.e vertical line so we use plt.axvline 
mean_score = df['Score'].mean()
median_score = df['Score'].median()
mode_score = df['Score'].mode()[0] # mode_score always returns a series AS maybe two or more columns may have tied for most frequent scores...
plt.axvline(mean_score, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean_score:.2f}')
plt.axvline(median_score, color='green', linestyle='dotted', linewidth=2, label=f'Median: {median_score:.2f}')
plt.axvline(mode_score, color='purple', linestyle='dotted', linewidth=2, label=f'Mode: {mode_score:.2f}')

# addimg frequency for every bar ...( we know counts list has count for every bar)

for i in range(len(counts)):
    # Only label bars that aren't empty
    if counts[i]>0:
        # bin_edges[i] is the left side, bin_edges[i+1] is the right side
        # We put the text in the middle of the bar

        x_pos = (bin_edges[i] + bin_edges[i+1]) / 2
        y_pos =  counts[i] + 0.1
        plt.text(x_pos, y_pos, str(int(counts[i])), ha='center', fontsize=9)


        


plt.legend()

plt.show()

print(counts)
print(bin_edges)
```

```python
# What is Bivariate Analysis?
# The name comes from "Bi" (meaning two) and "Variate" (meaning variable). Up until now, we have looked at your data columns one by one in isolation. Now, we are going to "marry" two columns to see if they have a relationship.

# In our dataset, the relationship to test is: Score vs. Popularity (Members).

# Insight we get ..


# The Goal: Finding "Correlation"
# We want to answer a question: Does being popula ( more members ) make an anime higher rated( better score ?

# Scenario A: If an anime has millions of members, does it automatically have a 9.0+ score? (This is a Positive Correlation).

# Scenario B: Are there "Hidden Gems" that have a 9.0+ score but very few members? (These are Outliers).

# Scenario C: Is there no relationship at all, meaning people watch anything regardless of the score?

# Positive correlation --> graph( data points not a line ) looks like y=mx as members increases score also increases ...
# Negative correlation --> y=-mx+c( data points ) , opp og positive one which shows increased members decreases score..
# No correlation       ---> scattered data points 
```

```python
# -------------SCATTER PLOT---------------#

# How it works (The Basics):We treat the Members as the X-axis (horizontal).We treat the Score as the Y-axis (vertical).
# Every single anime in your list becomes one dot on the graph.

# What to look for:

# The Trend: If the dots form a "cloud" that moves from the bottom-left to the top-right, it means more popular animes generally get higher scores.

# The Clusters: Do all the TV dots huddle together in one corner while the MOVIE dots sit somewhere else?

# The "Loner" Dots: If you see a dot far away from everyone else, that is an anime that is either "Overhyped" (High members, Low score) 
# or a "Secret Masterpiece" (Low members, High score).

```

```python
# now members count are in millions in some cases so we divide them in millions first ..

# y axis to start from just below min score i.e 8.6 and max to 9.2  i.e above max score 

#------------------------
# we need to make insights for questions like high member -->  high score , what is high , for this we divide graph into 4 quadrants 
# To make sense of where an anime stands, we need a "Center Point."
# The Step: We will prepare to draw a Horizontal Line at the Average Score and a Vertical Line at the Average Member Count.

# The Result: This divides your graph into 4 Quadrants.

# Top-Right: High Score + High Popularity (The "Blockbusters").

# Top-Left: High Score + Low Popularity (The "Hidden Gems").

# Bottom-Right: Low Score + High Popularity (The "Overhyped").

# Bottom-Left: Low Score + Low Popularity (The "Niche").
#--------------------------

# we want to get insights for every type so  , The Step: We prepare a Color Map. For example: TV = Blue, Movie = Red, OVA = Green
```

```python



# We divide by 1,000,000 to turn "1,500,000" into "1.5" (Millions)
# This makes the X-axis clean and readable.
x_popularity = df['Memb'] / 1000000 
y_score = df['Score']


color_map = {'TV': 'royalblue', 'MOVIE': 'crimson', 'OVA': 'limegreen', 'SPECIAL': 'orange'}

# We create a list of colors, one for every row in our data
# This "loops" through the Type column and picks the right color from our map.
colors = [color_map.get(t, 'gray') for t in df['Type']]

# 3. SET UP THE CANVAS (Large and Wide)
plt.figure(figsize=(10, 8))


# x: Popularity in Millions
# y: Exact Score
# c: The list of colors we just made
# s: Size of the dots (100 is a good "B.Tech" size)
# alpha: 0.6 makes dots semi-transparent so we can see overlaps
# edgecolor: Adds a thin border to each dot so they don't "bleed" together
plt.scatter(x_popularity, y_score, c=colors, s=100, alpha=0.6, edgecolor='white')

#
# axhline: Horizontal line for Average Score
# axvline: Vertical line for Average Popularity (in Millions)
plt.axhline(y_score.mean(), color='black', linestyle='--', alpha=0.5, label='Avg Score')
plt.axvline(x_popularity.mean(), color='black', linestyle='--', alpha=0.5, label='Avg Popularity')

# We zoom into the scores (8.6 to 9.2) so the dots aren't squashed
plt.ylim(8.65, 9.15) 

plt.title('Quality vs. Popularity: The Top 100 Landscape', fontsize=16, pad=20)
plt.xlabel('Popularity (Members in Millions)', fontsize=12)
plt.ylabel('Anime Score', fontsize=12)


# Since we used colors, we need to tell the reader what they mean.
# This part is a bit advanced, so we keep it simple:
from matplotlib.lines import Line2D
legend_elements = [Line2D([0], [0], marker='o', color='w', label=k,
                          markerfacecolor=v, markersize=10) for k, v in color_map.items()]
plt.legend(handles=legend_elements, title="Anime Type")

plt.grid(True, linestyle=':', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python

x_memb = df['Memb']
y_score = df['Score']

# (The global extremes)
max_score = y_score.max()
max_popular = x_memb.max()


plt.figure(figsize=(18, 10))
plt.scatter(x_memb, y_score, c=colors, s=100, alpha=0.6, edgecolor='white')
plt.xscale('log') 

for i in range(len(df)):
    
    name = df['Title'].iloc[i]
    score = df['Score'].iloc[i]
    memb = df['Memb'].iloc[i]
    
    # --- CATEGORY A: THE GLOBAL OUTLIERS ---
    # Highest Rated?
    if score == max_score:
        plt.text(memb, score + 0.005, f"BEST RATED: {name}", color='black')
    
    # Most Popular?
    if memb == max_popular:
        plt.text(memb, score + 0.012, f"MOST POPULAR: {name}", color='black')

    # --- CATEGORY B: THE RELATIONSHIP OUTLIERS (The Contradictions) ---
    # Underrated :  High Score (>9.0) but Low Popularity (<500k)
    if score > 9.00 and memb < 250000:
        plt.text(memb, score + 0.005, f"GEM: {name}", color='darkgreen', fontsize=9)
        
    # Overrated : Lower Score (<8.8) but High Popularity (>1.5M)
    if score < 8.8 and memb > 1500000:
        plt.text(memb, score + 0.005, f"HYPED: {name}", color='darkred', fontsize=9)


plt.axhline(y_score.mean(), color='gray', linestyle='--', alpha=0.5)
plt.axvline(x_memb.median(), color='gray', linestyle='--', alpha=0.5)

plt.title('Complete Outlier Analysis: Kings, Gems, and Giants', fontsize=16, pad=20)
plt.xlabel('Popularity (Log Scale of Members)', fontsize=12)
plt.ylabel('Quality Score', fontsize=12)
plt.grid(True, which="both", linestyle=':', alpha=0.3)

plt.show()
```
